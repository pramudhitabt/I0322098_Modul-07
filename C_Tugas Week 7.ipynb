```python
#Buatlah sebuah fungsi untuk membuat matrix berdasarkan input dari user

    #Tulis kode kamu disini

def buat_matriks():

    baris = int(input("Masukkan jumlah baris: "))
    kolom = int(input("Masukkan jumlah kolom: "))

    matriks = []

    for x in range(baris):
        baris_matriks = []
        for y in range(kolom):
            elemen = int(input("Masukkan elemen pada baris {} kolom {}: ".format(x+1, y+1)))
            baris_matriks.append(elemen)
        matriks.append(baris_matriks)

    return matriks

matriks = buat_matriks()
print("Matriks yang dihasilkan adalah: ")
for baris in matriks:
    print(baris)
```

```python
#Buatlah sebuah fungsi untuk mencari determinan matrix berdasarkan input matrix dari user

    #Tulis kode kamu disini

import numpy as np

def cari_determinan():
    
    baris = int(input("Masukkan jumlah baris: "))
    kolom = int(input("Masukkan jumlah kolom: "))

    matriks = []

    for x in range(baris):
        baris_matriks = []
        for y in range(kolom):
            elemen = float(input("Masukkan elemen pada baris {} kolom {}: ".format(x+1, y+1)))
            baris_matriks.append(elemen)
        matriks.append(baris_matriks)

    determinan = np.linalg.det(matriks)

    return determinan

determ = cari_determinan()
print("Determinan matriks: ", determ)
    
```

```python
#Buatlah fungsi untuk mengakalikan dua buah matrix berdasarkan input user menggunakan nested loop

    #Tulis kode kamu disini
    
def perkalian_matriks():
  
    baris1 = int(input("Masukkan jumlah baris matriks pertama: "))
    kolom1 = int(input("Masukkan jumlah kolom matriks pertama: "))

    matriks1 = []
    for x in range(baris1):
        baris_matriks1 = []
        for y in range(kolom1):
            elemen = float(input("Masukkan elemen pada baris {} kolom {}: ".format(x+1, y+1)))
            baris_matriks1.append(elemen)
        matriks1.append(baris_matriks1)

    baris2 = int(input("Masukkan jumlah baris matriks kedua: "))
    kolom2 = int(input("Masukkan jumlah kolom matriks kedua: "))

    matriks2 = []
    for x in range(baris2):
        baris_matriks2 = []
        for y in range(kolom2):
            elemen = float(input("Masukkan elemen pada baris {} kolom {}: ".format(x+1, y+1)))
            baris_matriks2.append(elemen)
        matriks2.append(baris_matriks2)

    if kolom1 != baris2:
        print("Perkalian matriks tidak dapat dijalankan. Jumlah kolom matriks pertama harus sama dengan jumlah baris matriks kedua.")
        return None

    matriks_hasil = []
    for x in range(baris1):
        baris_matriks_hasil = []
        for y in range(kolom2):
            elemen_hasil = 0
            for z in range(kolom1):
                elemen_hasil += matriks1[x][z] * matriks2[z][y]
            baris_matriks_hasil.append(elemen_hasil)
        matriks_hasil.append(baris_matriks_hasil)

    return matriks_hasil
    
hasil = perkalian_matriks()
if hasil:
    print("Hasil perkalian matriks adalah: ")
    for baris in hasil:
        print(baris)
```
